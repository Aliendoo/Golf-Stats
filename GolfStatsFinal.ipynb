## Building a multivariable model using the final variables from pgaTourData

```python
import numpy as np
import statsmodels.api as sm
```

```python
golfStats.columns.tolist()
```

```python
x_columns = golfStats.columns.tolist()
golfStats = golfStats.dropna()
y = golfStats["FinalScore"]
```

```python
x_columns.remove("FinalScore")
x_columns.remove("Round")
x_columns.remove("PuttsPerHole")
```

```python
def get_stats():
    x = golfStats[x_columns]
    results = sm.OLS(y, x).fit()
    print(results.summary())
get_stats()
```
